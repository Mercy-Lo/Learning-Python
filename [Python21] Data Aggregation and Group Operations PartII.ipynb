### 定義一個函式為最大值減去最小值。

```python
def peak_to_peak(arr):
    return arr.max()-arr.min()
```

### agg(): 可以用來執行聚合的函式。

```python
a.agg(peak_to_peak)
```

```python
a.describe()
```

### 欄方向的多功能運用

```python
tips=pd.read_csv('examples/tips.csv')
```

```python
tips[:6]
```

```python
tips['tip_pct']=tips['tip']/(tips['total_bill']-tips['tip'])
```

```python
tips[:6]
```

```python
a=tips.groupby(['day','smoker'])
```

```python
a_pct=a['tip_pct']
```

### agg(): 執行聚合函式。

```python
a_pct.agg('mean')
```

```python
a_pct.agg(['mean','std',peak_to_peak])
```

### agg((name,function)) : 可以加入tuple，自行定義該function是什麼名稱。

```python
a_pct.agg([('foo','mean'),('bar',np.std)])
```

### 也可以在agg內放入list，一口氣進行多個聚合函式運算。

```python
functions=['count','mean','max']
```

```python
result=a['tip_pct','total_bill'].agg(functions)
```

```python
result
```

```python
result['tip_pct']
```

```python
ftuples=[('Hello','mean'),('World',np.var)]
```

```python
a['tip_pct','total_bill'].agg(ftuples)
```

### 除了使用tuple，也可以使用dict，key就代表名稱，value就是函式功能。

```python
a.agg({'tip':np.max, 'size': 'sum'})
```

### tip_pct這個名稱內也可以同時包含非常多的聚合函式。

```python
a.agg({'tip_pct':['min','max','mean','std'],'size':'sum'})
```

### 預設值，分組依據的day, smoker會成為groupby的index。

```python
tips.groupby(['day','smoker']).mean()
```

### 如果不希望分組依據為Index，就可以寫上as_index=False。

```python
tips.groupby(['day','smoker'],as_index=False).mean()
```
